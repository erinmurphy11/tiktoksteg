```python
total_params = []

for first_c in [16, 32, 64, 128]:
    for max_c in [128, 256, 512, 1024, 2048, 4096]:
        for n_conv in range(1,8):
            for n_depthwise in range(1, n_conv):
                Hnet = HideNet(first_c=first_c, n_depthwise=n_depthwise, max_c=max_c, n_conv=n_conv)
                Rnet = RevealNet(nhf=first_c)
                
                Hnet_parameters = filter(lambda p: p.requires_grad, Hnet.parameters())
                Hnet_parameters = sum([np.prod(p.size()) for p in Hnet_parameters])

                Rnet_parameters = filter(lambda p: p.requires_grad, Rnet.parameters())
                Rnet_parameters = sum([np.prod(p.size()) for p in Rnet_parameters])
                
                print(f"first_c {first_c}, max_c {max_c}, n_conv {n_conv}, n_dw {n_depthwise}, {Hnet_parameters + Rnet_parameters}")
                total_params.append(Hnet_parameters + Rnet_parameters)
```
